```python
import matplotlib.pyplot as plt
import numpy as np
import cv2
```

```python
sample1 = np.random.randint(0, 255, (100, 100), dtype=np.uint8)
sample2 = np.random.randint(0, 255, (100, 100), dtype=np.uint8)

fig, ax = plt.subplots(1, 2)
ax[0].imshow(sample1, cmap='gray')
ax[1].imshow(sample2, cmap='gray')

```

```python
def get_arrays_by_plot(arr1, arr2, title1=None, title2=None):
    fig, ax = plt.subplots(1, 2)
    ax[0].imshow(arr1, cmap='gray')
    ax[1].imshow(arr2, cmap='gray')
    
    if title1:
        ax[0].set_title(title1)
    if title2:
        ax[1].set_title(title2)
        
    fig.canvas.draw()
        
    buf = fig.canvas.buffer_rgba()
    img = np.frombuffer(buf, np.uint8).reshape((buf.shape[0], buf.shape[1], 4))
    
    return img
```

```python
a = get_arrays_by_plot(sample1, sample2, 'sample1', 'sample2')

```
